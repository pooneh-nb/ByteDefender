```python
import utilities
from pathlib import Path
import pandas as pd
from beautifultable import BeautifulTable
import csv
project_dir = Path.home().cwd()
print(project_dir)
```

```python
output = utilities.get_directories_in_a_directory(Path.joinpath(project_dir, "server/output"))
dic = {}
api = 0
byte = 0
for site in output:
    site_name = site.split('/')[-1]
    dic[site_name] = {'API':False, 'Bytecode':False}
    files = utilities.get_files_in_a_directory(site)
    for file in files:
        if file.split('.')[-1] == 'json':
            dic[site_name]['API'] = True
            api += 1
        if file.split('.')[-1] == 'bin':
            dic[site_name]['Bytecode'] = True 
            byte += 1
```

```python
print(api, byte)
```

```python
df = pd.DataFrame.from_dict(dic, orient='index')
table = BeautifulTable()
table.columns.header = ['Site', 'API', 'Bytecode']

for index, row in df.iterrows():
    table.rows.append([index, *row])

print(table)
```

```python
# distribute tranco sites

def chunk_csv(input_file, lines_per_file):
    with open(input_file, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        chunk_number = 0
        current_line = 0
        for row in reader:
            web = [row[1]]
            if current_line % lines_per_file == 0:
                if current_line > 0:  # Close the previous file if it exists
                    output_file.close()
                chunk_number += 1
                out_dir = Path.joinpath(project_dir, 'tranco')
                output_path = out_dir / f'chunk_{chunk_number}.csv'
                output_file = open(output_path, 'w', newline='', encoding='utf-8')
                writer = csv.writer(output_file)
                writer.writerow(["website"])

            writer.writerow(web)
            current_line += 1
        if output_file and not output_file.closed:
            output_file.close()

# Usage
chunk_csv('tranco_1m.csv', 1000)
```
